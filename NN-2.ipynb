```
from keras import backend as K

def f1(y_true, y_pred):
    def recall(y_true, y_pred):
        """Recall metric.

        Only computes a batch-wise average of recall.

        Computes the recall, a metric for multi-label classification of
        how many relevant items are selected.
        """
        true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
        possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
        recall = true_positives / (possible_positives + K.epsilon())
        return recall

    def precision(y_true, y_pred):
        """Precision metric.

        Only computes a batch-wise average of precision.

        Computes the precision, a metric for multi-label classification of
        how many selected items are relevant.
        """
        true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
        predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
        precision = true_positives / (predicted_positives + K.epsilon())
        return precision
    precision = precision(y_true, y_pred)
    recall = recall(y_true, y_pred)
    return 2*((precision*recall)/(precision+recall+K.epsilon()))
```

```
#optuna based searching neural network parameters
import tensorflow as tf

import optuna

from tensorflow.keras.layers import Activation, Dropout, Dense, Conv2D, Flatten, Dropout, MaxPooling2D, BatchNormalization
from keras.utils import np_utils
import time
import tensorflow as tf

def plt_dynamic(x, vy, ty, ax, colors=['b']):
    """plot x wrt vy and x wrt ty"""
    ax.plot(x, vy, 'r', linestyle="solid", label="Validation Loss")
    ax.plot(x, ty, 'b', linestyle="dashed",label="Train Loss")
    plt.legend()
    plt.grid()
    #fig.canvas.draw()
    plt.show()

tr_x,tr_y,te_x,te_y=X_tr, y_tr, X_test, y_test

categorical_values=2 #this means the total labels we have so 2 because pos and neg
te_y=np_utils.to_categorical(te_y, categorical_values) 
tr_y=np_utils.to_categorical(tr_y, categorical_values) 
print(tr_x.shape,tr_y.shape,te_x.shape,te_y.shape)

#only softmax classifier

#some model parameters 
output_dim = tr_y.shape[1] #this is class labels 2
input_dim = tr_x.shape[1] #this is basically 300

# 1. Define an objective function to be maximized.
def objective(trial):

    #1. something that are kept static for a reason
    batch_size = 256 #for faster training and model prototyping
    nb_epoch=10 #epoch counter is set for 30 for faster training 
    batch_norm_after_layer=3 #we need to add batch norm after every 4th layer

    # 2. Suggest values of the hyperparameters using a trial object.
    n_layers = trial.suggest_int('n_layers', 1, 30) #gets the number of layers
    activation_list=["relu","selu","sigmoid","tanh","swish","softplus"]
    activation_type=trial.suggest_categorical('activation', activation_list)
    start=time.time()

    #initialize the model
    model = tf.keras.Sequential()
    
    for i in range(n_layers):
        #denotes hidden units in the ith layer
        num_hidden = int(trial.suggest_loguniform('n_units{}'.format(i), 4, 128))

        if(activation_type in ["sigmoid","tanh","swish","softplus"]):
          init_type="glorot_normal"
        else:
          init_type="he_normal"

        if(i==0): #basically first layer
          model.add(Dense(num_hidden, activation=activation_type, input_shape=(input_dim,)))
          continue
        
        #add the layer to the network
        model.add(Dense(num_hidden, activation=activation_type, kernel_initializer=init_type))

        #add a dropout rate 
        dropout_rate = trial.suggest_uniform('dropout_rate', 0.0, 1.0)
        model.add(Dropout(dropout_rate))

        if(i%batch_norm_after_layer==0):
          model.add(BatchNormalization())

    #add final softmax layer
    model.add(Dense(output_dim, activation='softmax'))

    #compile the model
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=[f1])

    #checkpoint the model
    checkpoint = tf.keras.callbacks.ModelCheckpoint(filepath="/content/drive/My Drive/results/best_nn.hdf5", monitor='val_loss', save_best_only=True, mode='min', verbose=1)
    callbacks_list = [checkpoint]

    #fit the model
    history = model.fit(tr_x, tr_y, batch_size=batch_size, epochs=nb_epoch, validation_data=(te_x, te_y),callbacks=callbacks_list) 
    model.summary()
    score = model.evaluate(te_x, te_y, verbose=0) 

    print('Test score:', score[0]) 
    print('Test accuracy:', score[1])

    end=time.time()
    print("the total time :",end-start)

    import matplotlib.pyplot as plt
    #get the plots
    fig,ax = plt.subplots(1,1)
    ax.set_xlabel('epoch') ; ax.set_ylabel('Categorical Crossentropy Loss')

    # list of epoch numbers
    x = list(range(1,nb_epoch+1))

    vy = history.history['val_loss'] #validation means this is the test loss 
    ty = history.history['loss'] #this is the train loss

    plt_dynamic(x, vy, ty, ax)

    return score[0]

# 3. Create a study object and optimize the objective function.
study = optuna.create_study(direction="minimize",study_name='NN_tuning', storage='sqlite:///nn.db', load_if_exists=True)
study.optimize(objective, n_trials=50,show_progress_bar=True,n_jobs=1)

df = study.trials_dataframe()
display(df)

optuna.visualization.plot_optimization_history(study)
```
